**BATTING DATASET DESCRIPTION**

---



1. **Player:** The name of the cricket player.

2. **Matches:** The total number of matches played by the player.

3. **Innings:** The total number of innings played by the player.

4. **Not Outs:** The number of times the player remained not out after completing an innings.

5. **Runs:** The total number of runs scored by the player across all matches.

6. **BF (Balls Faced):** The total number of balls faced by the player while batting.

7. **Highest Score:** The highest individual score (number of runs) achieved by the player in a single inning.

8. **Average:** The average number of runs scored by the player per dismissal (innings where the player got out).

9. **Strike Rate:** The strike rate of the player, calculated as (Runs/BF) * 100.

10. **Hundreds:** The number of centuries (individual innings with 100 or more runs) scored by the player.

11. **Fifties:** The number of half-centuries (individual innings with 50 to 99 runs) scored by the player.

12. **Ducks:** The number of times the player scored zero runs in an innings.

13. **Fours:** The total number of boundaries (four runs) scored by the player.

14. **Sixes:** The total number of sixes (six runs) scored by the player.

15. **Country:** The country to which the player belongs or represents in international cricket.

**BOWLING DATASET DESCRIPTION**


---


1. **Player:** The name of the cricket player.

2. **Span:** The time period (start and end years) during which the player participated in matches.

3. **Mat (Matches):** The total number of matches played by the player.

4. **Inns (Innings):** The total number of matches in which the player bowled.

5. **Overs:** The total number of overs bowled by the player.

6. **Mdns (Maidens):** The number of maidens bowled by the player. A maiden over is an over in which no runs are scored.

7. **Runs:** The total number of runs conceded by the player.

8. **Wkts (Wickets):** The total number of wickets taken by the player.

9. **BBI (Best Bowling in an Inning):** The date of best bowling performance by the player in a single inning.

10. **Ave (Average):** The average number of runs conceded per wicket taken by the player, calculated as Runs/Wkts.

11. **Econ (Economy Rate):** The economy rate of the player, indicating the average number of runs conceded per over bowled.

12. **SR (Strike Rate):** The bowling strike rate of the player, representing the average number of balls bowled per wicket taken.

13. **Five:** The number of times the player took five wickets in a single inning.

14. **Ten:** The number of times the player took ten wickets in a match.

15. **Ct (Catches):** The total number of catches taken by the player as a fielder.

16. **St (Stumpings):** The total number of stumpings made by the player as a wicketkeeper.

17. **Country:** The country to which the player belongs or represents in international cricket, indicating the national team they play for.

```python
def plot_wickets_pie(team_data, threshold=5):
    team_data['Player'] = team_data.apply(lambda row: row['Player'] if row['Wkts'] >= threshold else 'Others', axis=1)
    grouped_data = team_data.groupby('Player')['Wkts'].sum().reset_index()
    filtered_data = grouped_data[grouped_data['Wkts'] >= threshold]
    if len(grouped_data[grouped_data['Wkts'] < threshold]) > 1:
        others_row = pd.DataFrame({'Player': ['Others'], 'Wkts': [grouped_data[grouped_data['Wkts'] < threshold]['Wkts'].sum()]})
        filtered_data = pd.concat([filtered_data, others_row])
    plt.figure(figsize=(4,4))
    plt.pie(filtered_data['Wkts'], labels=filtered_data['Player'], autopct='%1.1f%%', startangle=90)
    plt.title(f'Wickets Distribution - {team_data["Country"].iloc[0]}')
    plt.show()
india_data = df[df['Country'] == 'IND']
plot_wickets_pie(india_data)
aus_data = df[df['Country'] == 'AUS']
plot_wickets_pie(aus_data)
nz_data = df[df['Country'] == 'NZ']
plot_wickets_pie(nz_data)
sa_data = df[df['Country'] == 'SA']
plot_wickets_pie(sa_data)
```

Mean Squared Error (MSE), Mean Absolute Error (MAE), and R-squared (R²) Metrics for different algorithms

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(features, indian_players['Runs'], test_size=0.2, random_state=42)
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)

mae_dt = mean_absolute_error(y_test, dt_predictions)
mse_dt = mean_squared_error(y_test, dt_predictions)
r2_dt = r2_score(y_test, dt_predictions)

rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)

mae_rf = mean_absolute_error(y_test, rf_predictions)
mse_rf = mean_squared_error(y_test, rf_predictions)
r2_rf = r2_score(y_test, rf_predictions)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)

mae_lr = mean_absolute_error(y_test, lr_predictions)
mse_lr = mean_squared_error(y_test, lr_predictions)
r2_lr = r2_score(y_test, lr_predictions)

print("Decision Tree Metrics:")
print("MAE: {:.2f}".format(mae_dt))
print("MSE: {:.2f}".format(mse_dt))
print("R-squared: {:.2f}".format(r2_dt))
print("\n")

print("Random Forest Metrics:")
print("MAE: {:.2f}".format(mae_rf))
print("MSE: {:.2f}".format(mse_rf))
print("R-squared: {:.2f}".format(r2_rf))

print("Linear Regression Metrics:")
print("MAE: {:.2f}".format(mae_lr))
print("MSE: {:.2f}".format(mse_lr))
print("R-squared: {:.2f}".format(r2_lr))
```
